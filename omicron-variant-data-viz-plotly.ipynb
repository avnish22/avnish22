# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import plotly.express as px
import plotly.graph_objects as go
from plotly.offline import download_plotlyjs, init_notebook_mode,plot
init_notebook_mode(connected=True)
import cufflinks as cf
cf.go_offline()
```

```python
df = pd.read_csv('../input/omicron-variant-of-covid19/covid-cases-omicron.csv')
```

# Read Data

```python
df.head()
```

```python
df.describe()
```

# Visualization

```python
fig = px.bar(df, x = 'Entity',y = 'Omicron_percentage',text = 'Omicron_percentage',color = 'Entity',
            title='Omicron percentage of countries')
fig.update_layout(uniformtext_minsize=10 , xaxis_tickangle=-45)
fig.update_traces(texttemplate='%{text:.2s}')
```

```python
fig = px.bar(df, x = 'Entity',y = 'Omicron_percentage',text = 'Omicron_percentage',color = 'Day',
            title='Omicron percentage of countries with datetime')
fig.update_layout(uniformtext_minsize=10 , xaxis_tickangle=-45)
fig.update_traces(texttemplate='%{text:.2s}')
```

# Omicron percentage in South Africa

```python
df_sa = df[df['Entity']=='South Africa']
df_sa
```

```python
fig = px.line(df_sa, x = 'Day',y = 'Omicron_percentage',
            title='Omicron percentage in South Africa')
fig
```
